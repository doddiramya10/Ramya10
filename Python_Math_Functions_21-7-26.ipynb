```python
#Demonstrate all assignment operators (=, +=, -=, *=, /=, //=, %= and **=) on a variable.
def function(operation):
    a=190
    if operation == "+=":
     a += 10
     print(a)
    elif operation == "-=":
      a -= 10  
      print(a)
    elif operation == "*=":
      a *= 5
      print(a) 
    elif operation == "/=":
      a /= 7
      print(a)
    elif operation == "//=":
      a //= 6
      print(a)
    elif operation == "%=":
      a %= 9
      print(a)
    elif operation =="**":
      a **= 3
      print(a)
    else:
      print("Invalid Operator")         
operation = input("Enter the operator: +=,-=,*=,/=,//=,%=,**")   
function(operation)


```
