```python
employees = [
    {"name": "Ram", "dept": "HR", "salary": 50000},
    {"name": "John", "dept": "HR", "salary": 60000},
    {"name": "Sita", "dept": "IT", "salary": 70000}
]
```

```python
#Create a dictionary of department-wise total salary
hr_total=0
it_total=0
for i in employees:
    if i["dept"]=="HR":
        hr_total+=i["salary"]
    elif i["dept"]=="IT":
        it_total+=i["salary"]
print({"HR":hr_total,"IT":it_total})           


```

```python
#7. Create a dictionary of department-wise average salary
hr_total=0
it_total=0
count=0
count1=0
for i in employees:
    if i["dept"]=="HR":
        hr_total+=i["salary"]
        count+=1
        average=hr_total/count
    elif i["dept"]=="IT":
        it_total+=i["salary"]
        count1+=1
print({"HR":average,"IT":it_total})
```
